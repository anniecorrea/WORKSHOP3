```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

read the data

```python
df_2015 = pd.read_csv('../data/2015.csv')
df_2016 = pd.read_csv('../data/2016.csv')
df_2017 = pd.read_csv('../data/2017.csv') 
df_2018 = pd.read_csv('../data/2018.csv')
df_2019 = pd.read_csv('../data/2019.csv')
```

EDA 2015

```python
df_2015.head(20)
```

```python
print("Filas y columnas:", df_2015.shape)
print("\nTipos de datos:\n", df_2015.dtypes)
print("\nValores nulos:\n", df_2015.isnull().sum())
print("\n duplicados:\n", df_2015.duplicated().sum())
```

```python
df_2015.describe()
```

```python
plt.figure(figsize=(12, 6))
region_counts = df_2015['Region'].value_counts()
sns.barplot(x=region_counts.index, y=region_counts.values, color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title('Number of Countries per Region (2015)')
plt.xlabel('Region')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()
```

```python
#Top 10 happiest countries by score
top_20_2015 = df_2015.nlargest(20, 'Happiness Score')
plt.figure(figsize=(10, 6))
sns.barplot(x='Happiness Score', y='Country', data=top_20_2015, color='lightgreen')
plt.title('Top 20 Happiest Countries (2015)')
plt.xlabel('Happiness Score')
plt.ylabel('Country')
plt.show()
```

```python
# Correlation matrix (use only numeric columns to avoid conversion errors)
numeric_df = df_2015.select_dtypes(include=[np.number])
if numeric_df.empty:
	print("No numeric columns available for correlation.")
else:
	corr_matrix = numeric_df.corr()
	plt.figure(figsize=(10, 8))
	sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
	plt.title('Correlation Matrix (2015)')
	plt.show()
```

```python
# plot all numeric column density plots 
cols = numeric_df.columns.tolist()
n = len(cols)
ncols = 3  # change this to adjust number of plots per row
nrows = math.ceil(n / ncols)

fig, axes = plt.subplots(nrows, ncols, figsize=(ncols * 5, nrows * 3.5))
axes = axes.flatten() if n > 1 else [axes]

for i, column in enumerate(cols):
    ax = axes[i]
    sns.kdeplot(numeric_df[column].dropna(), ax=ax, fill=True)
    ax.set_title(f'Density: {column} (2015)')
    ax.set_xlabel(column)
    ax.set_ylabel('Density')

# turn off any unused subplots
for j in range(n, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```

```python

# Tomamos solo columnas numricas
n = len(cols)
ncols = 3                      # Nmero de columnas en la cuadrcula
nrows = math.ceil(n / ncols)   # Calcula las filas necesarias

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, 4*nrows))
axes = axes.flatten()  # Para iterar fcilmente sobre los ejes

for i, col in enumerate(cols):
    series = df_2015[col].dropna()
    if not series.empty:
        axes[i].boxplot(series, vert=False)
        axes[i].set_title(f"{col}")
    else:
        axes[i].set_visible(False)

# Ocultar subplots vacos
for j in range(i+1, len(axes)):
    axes[j].set_visible(False)

fig.suptitle("Boxplots de variables numricas", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```
